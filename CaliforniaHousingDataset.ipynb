### Visualizing relationships

```python
fig, ax = plt.subplots(figsize=(12, 8))

plt.scatter(CA_housing_data['total_rooms'], CA_housing_data['median_house_value'])

plt.xlabel('Total rooms')
plt.ylabel('Median house value')
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

plt.scatter(CA_housing_data['housing_median_age'], CA_housing_data['median_house_value'])

plt.xlabel('Median age')
plt.ylabel('Median house value')
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

plt.scatter(CA_housing_data['median_income'] * 10000, CA_housing_data['median_house_value'])

plt.xlabel('Median income')
plt.ylabel('Median house value')
```

```python
CA_housing_data_corr = CA_housing_data.corr()

CA_housing_data_corr
```

```python
fig, ax = plt.subplots(figsize=(12, 10))

sns.heatmap(CA_housing_data_corr, annot=True)
```
